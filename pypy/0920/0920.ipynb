## 인접리스트

 [0, 1, 0, 1, 0],

<tip>
파이썬은 딕셔너리로도 구현할 수 있다.
graph = [
    '0': [1, 3],
    '1':[0, 2, 3, 4],
    '2':[1],
    '3':[0, 1, 4],
    '4':[1, 3],
]

메모리적으로 인접 행렬에 비해 효율적이다.

## DFS
    - 시작 정점의 한 방향으로 갈 수 있는 경로까지 깊이 탐색해 가다가 더 이상 없게 되면, 가장 마지막에 만났던 갈림길 간선이 있는 정점으로 되돌아와서 다른 방향의 정점으로 탐색을 계속 반복하여 결국 모든 정점을 순회하는 방법


## 서로소 집합들

### 서로소 집합(Disjoint-sets)
    - 서로 중복된 원소가 없는 집합들
    - 집합에 속한 하나의 특정 멤버를 통해 각 집합을 구분한다. 
    - 특정멤버=대표자    
    상호배타 집합 연산
    0. 집합생성(Make-Set(x))
    1. 대표자 저장 ( Union(x, y) 같은 그룹으로 묶기)
    2. 각 요소가 내가 속한 그룹의 대표자를 어떻게 찾을지 ?? ( Find-Set(x) )

    * 위를 Union-Find 알고리즘이라고 한다
### 상호배타 집합을 표현하는 방법
    - 연결리스트
    - 트리
