## 모델필드
1. CharField() : 글자 수 제약있는
2. TextField() : 없는
3. DatetimeField()
    - auto_now : 데이터가 저장될 때마다 자동으로 현재 날짜시간을 저장
    -auto_now_add : 데이터가 처음생성될 때만 자동으로 날짜시간 저장
