## 문제풀이 팁
1. 문제 해석
-문제에서 원하는 목표
-> 명확하게 문제를 파악
ex
<1486. 장훈이의 높은선반>
-서점
    - 높이가 B인 선반
    - 선반
        - 키가 큰 장훈이는 물건을 자유롭게 사용할 수 있다.
    - 점원(키 Hj)
        - 1명: 점원의 키 == 탑의 높이
        - 2명이상: 점원의 키 합 == 탑의 높이
        -> 선반보다 높거나 같을 경우 물건을 잡을 수 있다.
     - 높이가 B 이상인 탑 중에서, 높이가 가장 낮은 탑을 구해라.

-> backtracking 획기적으로 완전탐색의 경우의 수를 줄여보자.
