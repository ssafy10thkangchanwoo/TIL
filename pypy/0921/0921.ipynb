## 신장트리
    1. 모든 정점이 연결되어 있음.
    2. 싸이클이 존재하지 않는 부분 그래프 (간선 그래프 개수: N-1개)
    3. 한 그래프에서 여러개의 신장트리가 나올 수 있다.

## 최소신장트리(Minimum Spanning Tree)
    신장트리를 구성하는 간선들의 가중치합이 최소인 그래프


### MST 표현
    1. 갈 수 있는 곳들 중 제일 짧은 곳으로 가자, 모든 정점을 방문할 때까지
    2. 전체 간선들 중에서 제일 가중치가 적은 곳부터 선택, 사전에 가중치 정렬 필요

## PRIME 알고리즘
    1. 임의의 정점 하나를 선택하여
    2. 선택한 정점과 인접하는 정점들 중의 최소 비용 간선이 존재하는 정점 선택
    
    서로소(내가 방문한 점, 방문 안한 점) 정보를 유지

## Kruscal 
    1. 모든 간선들 중에서 비용이 작은 것들부터 고르자

## Dijkstra
    1. 시작 정점에서 거리가 최소인 정점을 선택해 나감.
    2. prime과 누적거리를 기준으로 삼는 것에 차이가 있음.
    3. 현재 시점 최단거리로 왔는데 다음 선택지에서 최단거리를 선택하면 다음 선택지에서도 최단거리 보장됨.

### 최단거리 문제 유형
    1. 특정 지점 -> 도착 지점까지 최단 거리 : 다익스트라
    2. 가중치에 음수가 포함 : 벨만포드
    3. 여러 지점 -> 여러 지점까지의 최단거리 : 플로이드-워샬, 
    시간제약 많으면 여러 지점에서 다익스트라 해도 됨
