## 알고리즘 설계 기법의 종류
1. 전체를 다 보자 (Brute Foce- 완전 탐색)
    1. 배열 :  for, while
    2. 그래프 : DFS, BFS

2. 상황마다 좋은 걸 고르자(Greedy- 탐욕)
    1. 규칙을 찾는 것
    2. 주의사항 : 항상 좋은 것을 뽑는 것이 최종 결과가 제일 좋은 것은 아니다.

3. 하나의 큰 문제를 작은 문제로 나누어 부분적으로 해결하자(Dynamic Programming)
    1. Memoization 기법을 활용
    2. 점화식(bottom-up), 재귀(top-down)

4. 큰 문제를 작은 문제로 쪼개서 해결하자(Divide and Conquer -분할정복)

5. 전체를 보되, 가능성 없는 것을 빼고 보자(백트랙킹)
    1. 가지치기

## 분할정복기법
### 분할정복 
    대표적인 알고리즘 사용 예: easy)거듭제곱,
    - 병합정렬: 여러개의 정렬된 자료의 집합을 병합하여 한 개의 정렬된 집합으로 만드는 방식
    -> 직접 구현할 일은 적다. 
       멀티쓰레드 개념과 매우 비슷
    - 퀵정렬
    -> 평균적으로 굉장히 좋음 (N log N)
    -> 특히 큰 데이터를 다룰 때 좋다.
    -> 단점: 역순정렬 등 최악의 경우 0(N^2)
    - 이진탐색
    
    Top-down 방식 
    시간복잡도 : O(n long n)
